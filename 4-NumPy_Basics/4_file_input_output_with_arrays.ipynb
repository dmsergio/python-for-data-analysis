### File I/O with arrays

```python
import numpy as np
```

Comúnmente la lectura y escritura de archivos se realiza con Pandas, aunque eso no quiere decir que NumPy no tenga sus propios métodos para tal fin

El método **save** guarda por defecto todo el contenido binario bruto del array de manera no comprimida en un fichero ".npy" 

```python
arr = np.random.randn(100)
np.save('array', arr)
```

```python
arr
```

El método **load** permite cargar el fichero

```python
arr_loaded = np.load('array.npy')
arr_loaded
```

El método **savez** permite guardar en disco más de un array, indicando cada uno de estos arrays con un keyword

```python
arr_a, arr_b = np.arange(10000), np.arange(10000)
```

```python
np.savez('two_arrays', a=arr_a, b=arr_b)
```

```python
arrays = np.load('two_arrays.npz')
arrays
```

```python
arrays['a'], arrays['b']
```

En arrays con gran cantidad de datos, para poder comprimirlo mejor sería conveniente usar el método **savez_compressed**

```python
arr_a1, arr_b1 = np.arange(10000), np.arange(10000)
np.savez_compressed('two_arrays_compressed', a=arr_a1, b=arr_b1)
```
